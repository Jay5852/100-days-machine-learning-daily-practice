```python
from sklearn.datasets import make_regression
import numpy as np

```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

# For b=0


```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b =0

loss_slope = -2 * np.sum(y - m*X.ravel() -b)
loss_slope
```

```python
#Calculating Learning Rate
lr =  0.1

step_size = lr * loss_slope
step_size
```

```python
#Calculating the new intercept
b = b -  step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

# For B = 100 (Away from the actual fit line)

### Starting with b = 100

```python
m = 78.35
b_initial = 100
lr = 0.1

y_pred_b100_init = ((m * X) + b_initial).reshape(4)

plt.figure(figsize=(10,6))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred_b100_init,color='orange',label='b = {}'.format(b_initial))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0 (previous run)')
plt.legend()
plt.title('Initial Prediction with b = 100')
plt.show()
```

#### Iteration 1 for b = 100

```python
loss_slope_b100 = -2 * np.sum(y - m*X.ravel() - b_initial)
print(f"Loss slope: {loss_slope_b100}")

step_size_b100 = lr * loss_slope_b100
print(f"Step size: {step_size_b100}")

b_new_1 = b_initial - step_size_b100
print(f"New b after iteration 1: {b_new_1}")

y_pred_b100_1 = ((m * X) + b_new_1).reshape(4)

plt.figure(figsize=(10,6))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred_b100_1,color='orange',label='b = {}'.format(b_new_1))
plt.plot(X,y_pred_b100_init,color='#A3E4D7',label='b = {} (initial b=100)')
plt.legend()
plt.title('Prediction after 1st Iteration (b_initial=100)')
plt.show()
```

#### Iteration 2 for b = 100

```python
loss_slope_b100_2 = -2 * np.sum(y - m*X.ravel() - b_new_1)
print(f"Loss slope: {loss_slope_b100_2}")

step_size_b100_2 = lr * loss_slope_b100_2
print(f"Step size: {step_size_b100_2}")

b_new_2 = b_new_1 - step_size_b100_2
print(f"New b after iteration 2: {b_new_2}")

y_pred_b100_2 = ((m * X) + b_new_2).reshape(4)

plt.figure(figsize=(10,6))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred_b100_2,color='orange',label='b = {}'.format(b_new_2))
plt.plot(X,y_pred_b100_1,color='#A3E4D7',label='b = {} (after 1st iter)')
plt.legend()
plt.title('Prediction after 2nd Iteration (b_initial=100)')
plt.show()
```

#### Iteration 3 for b = 100

```python
loss_slope_b100_3 = -2 * np.sum(y - m*X.ravel() - b_new_2)
print(f"Loss slope: {loss_slope_b100_3}")

step_size_b100_3 = lr * loss_slope_b100_3
print(f"Step size: {step_size_b100_3}")

b_new_3 = b_new_2 - step_size_b100_3
print(f"New b after iteration 3: {b_new_3}")

y_pred_b100_3 = ((m * X) + b_new_3).reshape(4)

plt.figure(figsize=(10,6))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred_b100_3,color='orange',label='b = {}'.format(b_new_3))
plt.plot(X,y_pred_b100_2,color='#A3E4D7',label='b = {} (after 2nd iter)')
plt.legend()
plt.title('Prediction after 3rd Iteration (b_initial=100)')
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1
epochs = 10

for i in range(epochs):
  loss_slope = -2 * np.sum(y- m*X.ravel() - b)

  step_size = (lr * loss_slope)

  b = b - step_size

  y_pred = m*X + b

  plt.plot(X,y_pred)


plt.scatter(X,y)
```
