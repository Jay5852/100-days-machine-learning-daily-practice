# 1.Extract Year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

# 2.Extract Month


```python
date['month_no'] = date['date'].dt.month

date.head()
```

# 3.Month Name

```python
date['month_name'] = date['date'].dt.month_name()

date.head()
```

# 4.Extract Exact Date

```python
date['exact_day'] = date['date'].dt.day
# date.drop('exact_date',inplace=True)
date.head()
```

```python
# Day of the week
date['day_of_week'] = date['date'].dt.dayofweek

date.head()
```

```python
# Day of the week
date['day_of_week_name'] = date['date'].dt.day_name()

date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['day_of_week_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week


date.head()
```

# Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

#  Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
today_year = today.year
today_month = today.month

date_year = date['date'].dt.year
date_month = date['date'].dt.month

months_passed = (today_year - date_year) * 12 + (today_month - date_month)

print(months_passed)
```
