```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/titanic.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title('Passangers Travelling with')
```

```python
#Extracting Numerical Part
df['Number_numerical'] = pd.to_numeric(df['number'],errors ='coerce',downcast='integer')
```

```python
#Extracting Categorial Part
df['Number_categorical'] = np.where(df['Number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
fig = df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
